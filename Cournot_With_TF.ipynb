```python
import tensorflow as tf
import numpy as np

def findq(q, v=14.0):
    tf.reset_default_graph()
    
    _q = tf.Variable(tf.random_uniform(shape=[1], minval=0, maxval=v-q, dtype=tf.float32), dtype=tf.float32)
    
    # 1. /  _q * (v - (q + _q))
    qa = tf.add(float(q), _q)
    cond = tf.reduce_mean(v - qa)
    up = tf.cond(cond > 0, lambda: tf.subtract(v, qa), lambda: tf.div(1., 0.02))
    uq = tf.div(1., tf.multiply(_q, up) + 1.)
    
    loss = uq ##tf.square(uq)
    opt = tf.train.GradientDescentOptimizer(1.)
    train = opt.minimize(loss)
    
    with tf.Session() as sess:
        init = tf.initialize_all_variables()
        
        sess.run(init)
        print ("_q initial:", sess.run(_q))
        
        for step in range(10000):
            sess.run(train)
            if step % 100 == 0:
                print ('\t', step, sess.run(_q), sess.run(uq), sess.run(up))
        
        return np.around(sess.run(_q))
    
if __name__ == "__main__":
    q = findq(0, v=12.)
    for i in range(100):
        print (i, q)
        if q > 12.:
            q = 12.
        if q < 0:
            q = 0
        q = findq(q, v=12.)
    
            
```
